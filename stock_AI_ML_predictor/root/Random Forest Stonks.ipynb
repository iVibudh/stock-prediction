```python
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df_stock_clean_rf = pd.read_csv("C:\\Users\\T1Vibudh\\Desktop\\GitHub\\Stonks\\stock_AI_ML_predictor\\root\\stock_data\\2021-04-24.df_stock_clean_rf.csv",  index_col=0,  encoding = 'cp1252')
print(len(df_stock_clean_rf))
df_stock_clean_rf.head()
```

```python
df_stock_clean_rf.columns
```

```python
df_stock_clean_rf.dropna(subset = ["market_cap"], inplace=True)
df_stock_clean_rf.dropna(subset = ["price_close_10day_100"], inplace=True)
df_stock_clean_rf.dropna(subset = ["price_close_10day_z"], inplace=True)
df_stock_clean_rf.dropna(subset = ["price_close_yesterday_close_100"], inplace=True)
df_stock_clean_rf.dropna(subset = ["price_close_yesterday_z"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Volume"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Volume_10day_100"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Volume_10day_z"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Volume_yesterday_z"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Dollar_Interest"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Dollar_Interest_10day_100"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Dollar_Interest_10day_z"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Dollar_Interest_yesterday_z"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Day_of_week"], inplace=True)
df_stock_clean_rf.dropna(subset = ["Sessions"], inplace=True)
print(len(df_stock_clean_rf))
```

```python
df_stock_clean_rf.columns
```

```python
Xdf_dataset = df_stock_clean_rf[['market_cap', 
                                'price_close_10day_100', 
                                'price_close_10day_z', 
                                'price_close_yesterday_close_100',
                                'price_close_yesterday_z', 
                                'Volume', 
                                'Volume_10day_100',
                                'Volume_10day_z', 
                                'Volume_yesterday_z', 
                                'Dollar_Interest',
                                'Dollar_Interest_10day_100', 
                                'Dollar_Interest_10day_z',
                                'Dollar_Interest_yesterday_z', 
                                'Day_of_week', 
                                'Sessions']]
print(len(Xdf_dataset))
Xdf_dataset.head()
```

```python
Y1df_dataset = df_stock_clean_rf[['percent_gain_5_today']]
print(len(Y1df_dataset))
Y1df_dataset.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xdf_dataset,
                                                    Y1df_dataset,
                                                    test_size = 0.25,
                                                    random_state = 19)
X_train.shape
X_test.shape
y_train.shape
y_test.shape
```

```python
len(y_test[y_test.percent_gain_5_today > 0])
```

```python
regressor = RandomForestRegressor(n_estimators=20, random_state=0)
#n_estimators - Number of trees in the forest 
#random_state - Controls both the randomness of the bootstrapping of the samples used when building 
              # trees (if bootstrap=True) and the sampling of the features to consider when looking 
              # for the best split at each node (if max_features < n_features)
# Ref -> https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html


regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
#y_pred.shape
#y_pred
y_predict = []
for y in y_pred:
    if y > 0.95:
        y_predict.append(1)
    else:
        y_predict.append(0)
#y_predict
```

```python
f_importance = regressor.feature_importances_
feature = []

for col in Xdf_dataset:
    feature.append(col)
    
df_f_importance = pd.DataFrame({'Feature_Name' :  feature, 
                                'Importance':  f_importance})
df_f_importance
```

```python
print(confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
print(accuracy_score(y_test, y_predict))
XYRdf_dataset = X_test
XYRdf_dataset['Y_test'] = y_test
XYRdf_dataset['y_score'] = y_pred
XYRdf_dataset['y_predict'] = y_predict
#XYRdf_dataset['dataID_pageNo' : dataID_pageNo]
len(XYRdf_dataset)
XYRdf_dataset.head()


#XYRdf_dataset.to_csv("F:/Environmental Baseline Data/Version 4 - Final/Support files/Random Forest/Results_Training_Set.csv")
```
